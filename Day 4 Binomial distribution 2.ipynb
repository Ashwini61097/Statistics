```python
# Enter your code here. Read input from STDIN. Print output to STDOUT

import math
def fact(n):
    fact = math.factorial(n)
    return(fact)


def comb(n, x):
    if(x == 0):
        re = 1
    else: 
        re = fact(n)/(fact(n-x)*fact(x))
    return(re)

def ans1(n, r, p, q):
    ans = 0
    for i in range(0, r+1):
        ansx = comb(n, i) * ((p)**i) * ((q)**(n-i))
        ans = ans + ansx
    return(ans)

def ans2(n, r, p, q):
    ans2_ = 0
    for i in range(0, r):
        ansy = comb(n, i) * ((p)**i) * ((q)**(n-i))
        ans2_ = ans2_ + ansy
        FAns = 1 - ans2_
    return(FAns)
        
perp, n = input().split()

n = int(n)
p = int(perp)/100
q = 1 - p
r = 2

x = n - r

# print(n, p, q, r, x)

Ans1 = ans1(n, r, p, q)
Ans2 = ans2(n, r, p, q)
print(round(Ans1, 3))
print(round(Ans2, 3))
```
